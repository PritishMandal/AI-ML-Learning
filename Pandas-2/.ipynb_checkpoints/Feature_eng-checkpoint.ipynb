```python
# Melt & Pivot

df = pd.DataFrame({
    "country":["USA","USA","India","India"],
    "Year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
    "tax":[10,20,30,40]
})
melted_df=df.melt(
    id_vars=["country","year"],
    value_vara=["sales","profit"],
    var_name="matrics",
    value_name="value"
)
original = melted_df.pivot(
    index=["country","year"],
    colums="metrics"
    values="value"
)
```

```python
#Merging & Joining Data

df_customers = pd.DataFrames({
    "customer_id":[1,2,3,4],
    "name"["Adam","Bob","Charlie","Dave"]
})
df_orders = pd.DataFrames({
    "order_id":[101,102,103,104],
    "customer_id":[2,1,3,5],
    "amount":[250,120,300,180]
})
pd.merge(,df_customers,df_orders,on="customer_id")#innner join
pd.merge(,df_customers,df_orders,on="customer_id",how="left")#left join
pd.merge(,df_customers,df_orders,on="customer_id",how="right")#right join
pd.merge(,df_customers,df_orders,on="customer_id",how="outer")#outer join

```
