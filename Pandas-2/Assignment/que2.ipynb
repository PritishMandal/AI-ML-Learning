```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(df.head())
```

```python
df_A = df[['Name', 'Sex', 'Age', 'Fare', 'Survived']]
print(df_A.head())
```

```python
df_B = df[(df['Sex'] == 'female') & (df['Fare'] > 30)]
print(df_B)
```

```python
df_C = df.groupby('Pclass').agg({
    'Survived': 'mean',
    'Fare': 'mean',
    'Age': 'mean'
})

print(df_C)
```
