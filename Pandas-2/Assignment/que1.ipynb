```python
import pandas as pd

# Load dataset
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)

# Display first 10 rows
print(df.head(10))
```

```python
# Shape & Datatypes
print("Shape:", df.shape)
print("\nDatatypes:\n", df.dtypes)

# Summary statistics
print("\nSummary Statistics:\n", df.describe())
```

```python
filtered_df = df[
    (df['petal_length'] > 4.5) & 
    (df['species'] == "virginica")
]

print(filtered_df)
```

```python
grouped = df.groupby('species').agg({
    'sepal_length': 'mean',
    'petal_width': 'max',
    'sepal_width': 'std'
})

print(grouped)
```

```python
# Create new column
df['petal_ratio'] = df['petal_length'] / df['petal_width']

# Average petal_ratio per species
ratio_avg = df.groupby('species')['petal_ratio'].mean()

print(ratio_avg)
```
