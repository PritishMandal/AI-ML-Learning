```python
import json
```

```python
# Laosa the data
def load_data(filename):
    with open(filename,"r") as f :
        data = json.load(f)

    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
#function to clean the data
def clean_data(data):
    text_to_num = {"one" : 1, "two" : 2, "three" : 3, "four": 4, "five": 5}
    
    unique_users = set()
    cleaned_data = []
    
    for user in data:
        #clean rating - data consistent 
        raw_rating = str(user['rating']).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user['rating'] = raw_rating

        #handle missing vals
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        #hdeduplication
        if(user['name'].strip() in unique_users):
            continue
        unique_users.add(user['name'])
        cleaned_data.append(user)
        
    return cleaned_data
```

```python
data = clean_data(data)
print(data, len(data))
```

```python
# GET meaning insights from data
def get_insights(data):
    #avg rating
    tot_ratings = 0
    for user in data:
        tot_ratings += float(user['rating'])
    
    print(f"average product rating = {tot_ratings/len(data)}")

    #percentage of users with poor rating
    poor_ratings = 0
    for user in data:
        if(float(user['rating']) < 3.0):
            poor_ratings += 1
    
    print(f"% users with poor rating = {poor_ratings/len(data)*100}%")
```

```python
get_insights(data)
```

```python
#Build recommendation feature
def get_recommendations(data):
    recommendations = []
    
    for user in data:
        curr_recomm = {}
        curr_recomm['name'] = user['name']
        if(float(user['rating']) >= 4):
            curr_recomm['brand'] = "Apple"
        else:
            curr_recomm['brand'] = "Samsung"

        recommendations.append(curr_recomm)
    return recommendations
```

```python
get_recommendations(data)
```
