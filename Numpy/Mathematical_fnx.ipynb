```python
import numpy as np
```

```python
#Mathematical fnx

#aggregation 

arr = np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))
print(np.argmax(arr))
```

```python
#power fnx
arr = np.array([1,2,3,4,5])

print(np.square(arr))
print(np.sqrt(arr))
print(np.square(arr))
```

```python
#log

arr = np.array([1,2,3,4,5])

print(np.log(arr))
print(np.log2(arr))
```

```python
#rounding

print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))

arr = np.array([1,2,-8,3,4,-5,2,-5])
print(np.unique(arr))
print(np.sort(arr))
print(np.abs(arr))
```
